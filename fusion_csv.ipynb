### Data

```python
import numpy as np
import pandas as pd
import urllib
import io
import requests
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18000101-123954.csv')
df1 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18000101-151142.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18000101-173005.csv')
df3 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18000102-131053.csv')
df4 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18000101-112901.csv')
df5 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18000102-163308.csv')
df6 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18000104-114425.csv')
df7 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18000102-155538.csv')
df8 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18010206-154456.csv')
df9 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18010208-111625.csv')
df10 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18010213-121039.csv')
df11 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18010218-134025.csv')
df12 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18010418-110329.csv')
df13 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18010504-115138.csv')
df14 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18010501-163143.csv')
df15 = pd.read_csv('https://raw.githubusercontent.com/red-one-dataviz/fil_rouge/plotly/Data/FL_18010426-103237.csv')
```

### Fusion des fichiers csv en 1 seul csv

```python
csv_stack = pd.concat([df,df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13,df14,df15], axis=0)
csv_stack.head()
```

```python
csv_stack = csv_stack.drop(['date time', 'take off switch', 'ind. air speed', 'oat', 'n1 1', 'n2 1', 'nr', 'tot 1', 'power'], axis=1)
```
